```python
import pandas as pd
```

```python
data = [[1, 'a@b.com'], [2, 'c@d.com'], [3, 'a@b.com'], [4, 'a@b.com']]
person = pd.DataFrame(data, columns=['id', 'email']).astype({'id':'Int64', 'email':'object'})
```

```python
person = person[person.duplicated('email')].rename(columns={'email':'Email'})
```

```python
person
```

```python
person = person.drop_duplicates(subset=['Email']).drop('id', axis=1)
```
