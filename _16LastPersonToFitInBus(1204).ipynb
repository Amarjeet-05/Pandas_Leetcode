```python
import pandas as pd
```

```python
data = [[5, 'Alice', 250, 1], [4, 'Bob', 175, 5], [3, 'Alex', 350, 2], [6, 'John Cena', 400, 3], [1, 'Winston', 500, 6], [2, 'Marie', 200, 4]]
queue = pd.DataFrame(data, columns=['person_id', 'person_name', 'weight', 'turn']).astype({'person_id':'Int64', 'person_name':'object', 'weight':'Int64', 'turn':'Int64'})
```

```python
queue
```

```python
queue = queue.sort_values(by="turn")
queue
```

```python
sum = 0
item = []
for i in queue['weight']: 
    sum = sum + i
    item.append(sum)
# here we can also use cumsum().  (queue['cumweight'] = queue['weight'].cumsum())
```

```python
queue['Total Weight'] = item
```

```python
queue[queue['Total Weight'] <= 1000].tail(1)[['person_name']]
```
