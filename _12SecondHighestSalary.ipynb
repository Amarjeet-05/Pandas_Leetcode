```python
import pandas as pd
import numpy as np
```

```python
data = [[1, 100],[2,100],[3,50]]
employee = pd.DataFrame(data, columns=['id', 'salary']).astype({'id':'int64', 'salary':'int64'})
```

```python
employee
```

```python
df = employee['salary'].drop_duplicates().nlargest(2)
```

```python
df
```

```python
df = pd.DataFrame(df)
```

```python
df
```

```python
df = df[df['salary'] < df['salary'].max()]
if df['salary'].empty == True:
    df = pd.DataFrame({'salary' : [None]})
```

```python
df.rename(columns={'salary' : 'SecondHighestSalary'})
```
