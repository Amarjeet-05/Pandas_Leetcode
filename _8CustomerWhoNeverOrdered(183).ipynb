```python
import pandas as pd
```

```python
data = [[1, 'Joe'], [2, 'Henry'], [3, 'Sam'], [4, 'Max']]
customers = pd.DataFrame(data, columns=['id', 'name']).astype({'id':'Int64', 'name':'object'})
data = [[1, 3], [2, 1]]
orders = pd.DataFrame(data, columns=['id', 'customerId']).astype({'id':'Int64', 'customerId':'Int64'})
```

```python
customers
```

```python
orders
```

```python
df = customers.merge(orders, left_on='id', right_on='customerId', how='outer', indicator=True).query("_merge != 'both'").drop(columns='_merge')
```

```python
df['Customers'] = df['name']
```

```python
df = df[['Customers']]
```
