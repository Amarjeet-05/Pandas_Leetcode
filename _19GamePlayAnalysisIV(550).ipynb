```python
activity = pd.read_csv("player_activity_full.csv",parse_dates=["event_date"])
```

```python
activity
```

```python
activity = activity.sort_values(by='event_date')
```

```python
count = 0
id = {}
for i, act in activity.iterrows():
    if act['player_id'] not in id.keys():
        id[act['player_id']] = act['event_date']

    else:
        if (act["event_date"] + pd.Timedelta(days=-1)) == id[act['player_id']]: 
            count += 1
    
```

```python
factor = count/len(id)
```

```python
round(factor, 2)
```
