```python
import pandas as pd
```

```python
data = [[1, 1], [2, 1], [3, 1], [4, 2], [5, 1], [6, 2], [7, 2]]
logs = pd.DataFrame(data, columns=['id', 'num']).astype({'id':'Int64', 'num':'Int64'})
```

```python
logs
```

```python
cn = []
store = logs.iloc[0, 0]
count = 0
for i in logs['num']:
    if(i == store):
        count += 1
        if(count >= 3):
            cn.append(i)
    else:
        count = 0
        store = i
```

```python
df = pd.DataFrame(cn, columns=['ConsecutiveNums'])
```
