```python
df = pd.merge(emp, dept, left_on='departmentId', right_on='id', suffixes=('_emp', '_dept'))
```

```python
df
```

```python
df2 = df.drop_duplicates(subset=['salary','name_dept'])
df2
```

```python
top_3 = df2.groupby('name_dept')['salary'].nlargest(3).reset_index()
```

```python
top_3
```

```python
df = pd.merge(df, top_3, on=['name_dept', 'salary'])
```

```python
df.rename(columns={'name_dept': 'Department','name_emp': 'Employee','salary': 'Salary'})[['Department', 'Employee', 'Salary']]
```
